```python
#Tugas2

print('Ini adalah program sederhana untuk menghitung luas bangun datar ')
print("Pilih menu angka 1-3 : \n1. Persegi\n2. Lingkaran\n3. Segitiga")

pilihmenu = int(input("silahkan pilih menu dengan mengetikan angka 1-3 \n="))

match pilihmenu :
     case 1:

          print("ini adalah menu untuk menghitung luas persegi")
          sisi = int(input("Silahkan masukan nilai yang mau di hitung \n="))
          hitung = sisi * sisi
          print(f"Luas persegi adalah : {hitung}")

     case 2:

          print("ini adalah menu untuk menghitung luas lingkaran ")
          jarijari = int(input("Silahkan masukan nilai yang mau di hitung \n="))
          hitung = 22/7 * jarijari * jarijari
          print(f"Luas lingkaran adalah : {hitung}")

     case 3:

          print("ini adalah menu untuk menghitung luas segitiga ")
          alas = int(input("Silahkan masukan nilai yang alas \n="))
          tinggi = int(input("Silahkan masukan nilai yang alas \n="))
          hitung = 1/2 * alas * tinggi
          print(f"Luas segitiga adalah : {hitung}")

     case _:

          print("Pilihan tidak valid")
```
